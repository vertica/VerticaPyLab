````python
# Defining a function for multiple choice widgets

def create_multipleChoice_widget(options, correct_answer):
    if correct_answer not in options:
        options.append(correct_answer)
    
    correct_answer_index = options.index(correct_answer)
    
    radio_options = [(words, i) for i, words in enumerate(options)]
    alternativ = widgets.RadioButtons(
        options = radio_options,
        description = '',
        disabled = False
    )
    
    description_out = widgets.Output()
#    with description_out:
#        print(description)
        
    feedback_out = widgets.Output()

    def check_selection(b):
        a = int(alternativ.value)
        if a==correct_answer_index:
            s = widgets.HTML('<div class="correct-prompt">Correct!</div>')
        else:
            s = widgets.HTML('<div class="incorrect-prompt">Try Again!</div>')
        with feedback_out:
            clear_output()
            display(s)
        return
    
    check = widgets.Button(description="submit")
    check.on_click(check_selection)
    
    
    return widgets.VBox([description_out, alternativ, check, feedback_out])

def create_numeric_widget(correct_answer):
    
    #correct_answer_index = options.index(correct_answer)
    
    #radio_options = [(words, i) for i, words in enumerate(options)]
    alternativ = widgets.Text(
        #options = radio_options,
        #description = '',
        disabled = False
    )
    
    description_out = widgets.Output()
#    with description_out:
#        print(description)
        
    feedback_out = widgets.Output()

    def check_selection(b):
        try:
            a = int(alternativ.value)
        except ValueError:
            a = float(alternativ.value)
        if a==correct_answer:
            s = widgets.HTML('<div class="correct-prompt">Correct!</div>')
        else:
            s = widgets.HTML('<div class="incorrect-prompt">Try Again!</div>')
        with feedback_out:
            clear_output()
            display(s)
        return
    
    check = widgets.Button(description="submit")
    check.on_click(check_selection)
    
    
    return widgets.VBox([description_out, alternativ, check, feedback_out])
````

### Model Fitting and Analysis

Use the following to fit the model:

```Python
from verticapy.learn.linear_model import LinearRegression
model = LinearRegression(name = "public.LR_winequality")
model.fit(data,["TV","Radio","Newspaper"],"Sales")
```

The model parameters shown below are automatically displayed:

<a id='image_1_cell'></a>


So far, the data suggests the following:
- The high p-value of __Newspaper__ suggests that it does not have a strong (linear) relationship with __Sales__.
- __TV__ and __Radio__ have p-values of around 0, which suggests that they strongly affect __Sales__.
- The coefficient of __Radio__ is much higher than that of __TV__, suggesting that a change in __Radio__ results in a greater change in __Sales__ than would be seen from a similar change in __TV__.
- The coefficients of __TV__ and __Radio__ are both positive, so they are positively correlated.

There are several metrics you can use to assess the accuracy of the model, such as AIC and R-squared. The `regression_report()` function returns a table with all such metrics: 

```Python
model.regression_report()
```

The combination of high adjusted R-squared (__r2_adj__) and low Root-mean-square error (__root_mean_squared_error__) confirms that the model is very accurate.

### Model Modifications

<a id='h3_cell'></a>

You can sometimes improve the accuracy of a model by making modifications to it. One such modification is removing statistically insignificant features. In the previous example, the p-value of __Newspaper__ was [quite high](#image_1_cell), indicating a low statistical significance. To remove __Newspapers__ and check the accuracy of the new model:

```Python
model.fit(data,["TV","Radio"],"Sales")
model.regression_report()
model.score("r2adj")
```



While removing the __Newspaper__ feature does not significantly improve the adjusted R-squared value (the accuracy) of the model, the lack of any meaningful change in R-squared confirms that __Newspaper__ advertisements have no noticeable effect on __Sales__. 

Another thing worth investigating is the effects of using different advertising mediums together. For example, how does advertising on both __TV__ and __Radio__ affect __Sales__? To test this, the following example adds an interaction term $Radio \times TV$:

```Python
data.eval(name = "TVnRadio", expr = "TV*Radio")
```

To verify how __TVnRadio__ affects sales, calculate the p-value:

```Python
model.fit(data,["TV","Radio","TVnRadio"],"Sales")
```

The __TVnRadio__ interaction term has a very small p-value, suggesting that it has a strong effect on __Sales__. To verify this, check the accuracy of this new model with `score()`:

```Python
model.score("r2adj")
```

The adjusted R-squared value has jumped to 0.967, so the new model is very accurate.

Another way to potentially improve model accuracy is by including non-linear terms. For example, if you observed that the relationship between __Radio__ and __Sales__ is non-linear, then you could add $Radio^2$ as a variable.

The following example adds a __Radio_2__ variable to test this:

```Python
data.eval(name = "Radio_2", expr = "Radio*Radio")
model.fit(data,["TV","Radio","Radio_2"],"Sales")
```

Unfortunately, the __Radio_2__ variable produces a high p-value, indicating that the relationship between __Radio_2__ and __Sales__ is weak, so you should discard this term.

The previous examples yielded useful results and an improved model, but they still came down to trial and error based on R-squared, making it infeasible for models with many features. To overcome this, VerticaPy offers more sophisticated methods for finding the best model, such as step-wise linear regression, which adds more features and tries various combinations, and scores them based on the Bayesian Information Criterion (BIC).

The following example uses this method to generate several models with various __bic__ values (lower is better/more accurate):

```Python
features=data.get_columns()
features.remove('"Sales"')
stepwise(model,
         input_relation = data, 
         X =features, 
         y = "Sales",
         direction = "forward",)
```

These results show that a combination of __TV__, __TVnNewspaper__ (an interaction term), __Radio__, and __Newspaper__ produces the most accurate model.
