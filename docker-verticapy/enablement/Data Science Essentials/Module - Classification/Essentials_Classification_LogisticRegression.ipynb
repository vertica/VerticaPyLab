### Introduction to classification


When the feature of interest is qualitative or categorical (as opposed to quantitative), you should use classification algorithms. In such cases, you cannot use linear regression.

Some examples of use cases where classification is applicable include: 

- A. Determining whether a borrower will default on their debt.
- B. Identifying the type of illness of a patient.
- C. Determining whether an email is spam.

There are two types of classification:
- <b>Binary:</b> Problems where the answer is one of "Yes" or "No." Examples A and C are examples of this type.
- <b>Multinomial:</b> Problems where there are more than two options for the feature of interest. Example B is an example of this type.

```python
import numpy as np
from matplotlib import pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2)


data = np.array([
    [1, 2],
    [2, 3],
    [2.5, 6],
    [1.5,4],
])
x, y = data.T
ax1.scatter(x,y)

data = np.array([
    [3, 2],
    [3.5, 3],
    [3.2, 3],
    [4,2],
])
x, y = data.T
ax1.scatter(x,y,marker='p')

data = np.array([
    [5, 1],
    [6, 4],
    [5.6, 5],
    [7,4],
])
x, y = data.T
ax1.scatter(x,y,marker='s')
ax1.axvline(x = 2.7, color = 'b',linestyle="--")
ax1.axvline(x = 4.5, color = 'b',linestyle="--") 
ax1.set_ylabel('X2')
ax1.set_xlabel('X1')
ax1.set_title('Multi-class')
data = np.array([
    [1, 2.5],
    [2, 3.1],
    [2.5, 5.3],
    [1.5,4],
])
x, y = data.T
ax2.scatter(x,y)

data = np.array([
    [3, 2.4],
    [3.5, 2.7],
    [3.2, 5],
    [4,2],
])
x, y = data.T
fig.subplots_adjust(top=1)
fig.subplots_adjust(right=1.5)
ax2.scatter(x,y,marker='p')
ax2.axvline(x = 2.7, color = 'b',linestyle="--")
ax2.set_xlabel('X1')
ax2.set_ylabel('X2')
ax2.set_title('Binary')
#B=plt.show()

```

In general, the underlying goal of classification problems is determining the probability that a data point falls into a particular category, and then assigning that datapoint a label based on whether the probability exceeds a given threshold.

There are several methods for calculating these probabilities. Some methods include:
- Logistic regrsesion
- Linear discriminant analysis
- k-nearest neighbors
- Trees
- Random forest

### Limitations of Linear Regression

At first glance, it seems like you could use linear regression and encode the category to find its probability. For example, for binary classification, you could encode a "Yes" as 1 and a "No", as 0, train the model like any other linear regresson model, and then apply the threshold to classify the result:

 $p(X)= \beta_{0}+\beta_{1}X$

One issue with this approach is that some of the output probabilities could be greater than 1 or smaller than 0, making them difficult to interpret. Aside from that, this approach could work, and the result for using linear regression for a binary classification use case is identical to linear discriminant analysis. However, this method is still less efficient than logistic regression, which will be covered later in the module.

However, if the problem involves more than two classes, then using linear regression in this way will produce errors. This is because by encoding multiple categories into 0, 1, 2, 3, and so on, we also imply a relationship: that a category is semantically "between" another. That is, using linear regression in this way implies that category '1' is between categories '0' and '2,' which could be inaccurate or completely irrelevant information depending on the original problem.

### Logistic Regression

The main goal behind classification is to calculate the probability that a data point falls in a particular category. Logistic regression actually has this same goal.

To avoid the aforementioned issues of applying linear regression to a classification problem, the probability equation is modified as follows:

\begin{equation}
p(X)= \Large\frac{ \mathrm{e}^{\beta_{0}+\beta_{1}X}} {1+\mathrm{e}^{\beta_{0}+\beta_{1}X}}
\end{equation}

This equation always produces an S-shaped curve and ensures that all the values of probability will be between 0 and 1.

```python
import math
x=np.array(list(range(0,100)))
B0=-5
B1=0.1
plt.plot(x,np.exp(B0+B1*x)/(1+np.exp(B0+B1*x)))
plt.title('S-shaped curve')
plt.xlabel('X')
plt.ylabel('Probability')
```

Rearranging the equation:

\begin{equation}
\frac{p(X)}{1+p(X)}\ = \mathrm{e}^{\beta_{0}+\beta_{1}X}
\end{equation}

Taking the logarithm of both sides, you can make the right-hand side linear:

\begin{equation}
\mathrm{log}{(\frac{p(X)}{1+p(X)}})\ = \beta_{0}+\beta_{1}X
\end{equation}

These coefficients are not linearly related with the probabilities, so making inferences from the coefficients from the changes in $X$ and their results on $Y$ is not as straightforward as it would be with linear regression. Still, the signs of the beta terms indicate a positive or negative correlation. For example, a positive value for $\beta_{1}$ indicates that an increase in $X$ will be associated with increasing $p(X)$. 

The next step is to find the coefficients from the training data, typically with maximum likelihood estimation (MLE). MLE is used to fit many non-linear models.

\begin{equation}
  l(\beta_{0},\beta_{1})= \normalsize \prod_{i:y_{i}=1}   p(x_{i}) \prod_{i':y_{i'}=0} (1-p(x_{i'}))
\end{equation}

MLE finds the coefficients $\beta_{0}$ and $\beta_{1}$ such that the $p(X)$ is close to 1 for the values with "Yes" label and close to 0 for the values with "No" label.

The discussion so far has only concerned a single predictor (that is, $X$ has only one dimension), but this method can be applied and extended to multiple dimensions.
